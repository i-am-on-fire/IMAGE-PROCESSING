```python
import numpy as np
```

```python
train_semanticfeatures = np.load(r'FeatureFiles/train_semanticfeatures.npy')
test_semanticfeatures = np.load(r'FeatureFiles/test_semanticfeatures.npy')
```

```python
train_semanticfeatures[0]
```

```python
def semantic_normalizer(semantic_data):
    img_num = 0
    for data in semantic_data:
        norm = 0
        for ele in data:
            norm += ele**2
        norm = norm**0.5
        
        for i in range(len(data)):
            data[i] /= norm
            
        img_num += 1
        if img_num % 100 == 0:
            print(f'Done for {img_num} images')
```

```python
semantic_normalizer(train_semanticfeatures)
print('Done for train')
semantic_normalizer(test_semanticfeatures)
print('Done for test')
```

```python
np.save(r'FeatureFiles/train_semanticfeatures_normalized.npy', train_semanticfeatures)
np.save(r'FeatureFiles/test_semanticfeatures_normalized.npy', test_semanticfeatures)
```
